```julia
using Plots, Interact

xg = range(-2, 2, length=50)
yg = range(-2, 2, length=50)
@manipulate for  = range(-2, 2, length=7),
                 = range(-2, 2, length=5),
                 = range(-2, 2, length=5),
                b1 = range(-2, 2, length=5),
                b2 = range(-2, 2, length=5)
    f(x,y) =  * x^2 +  * y^2 +  * x * y - b1 * x - b2 * y
    surface(xg, yg, f)
end
```

```julia
using Plots, Interact

xg = range(-2, 2, length=50)
yg = range(-2, 2, length=50)

f(x,y) = 6 * x^2 + 2 * y^2 - 3 * x * y - 1.7 * x - 0.8 * y

@manipulate for  in range(0, 2, length=360),
                t1 in xg, t2 in yg
    transl = [t1; t2]
    rot = [cos() sin(); -sin() cos()]

    g(x,y) = f((rot * ([x;y] + transl))...)

    contour(xg, yg, g)
end
```

```julia
@manipulate for  = range(-2, 2, length=5),
                 = range(-2, 2, length=5),
                b = range(-2, 2, length=5)
    f(x,y) =  * x^2 +  * y^2 - b * x
    surface(xg, yg, f)
end
```
