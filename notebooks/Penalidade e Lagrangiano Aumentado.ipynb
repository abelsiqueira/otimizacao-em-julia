```julia
using LinearAlgebra

xg = range(-1.5, 1.5, length=200)
yg = range(-1.5, 1.5, length=200)

f(x) = x[1]^2 + 4 * x[2]^2
f(x,y) = f([x;y])
# c(x) = [x[1]^2 + x[2]^2 - 1]
c(x) = [x[1] + x[2] - 1]

contour(xg, yg, f, leg=false, levels=50)
g = range(0, 2, length=60)
# plot!(cos.(g), sin.(g), c=:black, lw=2)
plot!(xg, 1 .- xg, c=:black, lw=2)
# g = exp.(range(log(1e-4), log(1e4), length=50))
# x = 4 * g ./ (8 .+ 5g)
# y = g ./ (8 .+ 5g)
# plot!(x, y, c=:red, m=(1))
# scatter!([4/5], [1/5], c=:blue)

 = 1.0
x = [-0.4; 1.0]
m = 1

for k = 1:20
    scatter!([x[1]], [x[2]], c=:red, ms=2)
    cx = c(x)
    gx = f(x)
    Ax = J(x)
    Bx = H(x) +  * sum(Hc(x, i) * cx[i] for i = 1:m)
    x = -(Bx +  * Ax' * Ax) \ (gx +  * Ax' * cx)
    plot!(x[1] .+ [0, x[1]], x[2] .+ [0, x[2]], c=:red, l=:arrow)
    x += x
    y =  * c(x)
     = max(1.0, 2)
    println("$k  $(norm(cx)) $y  $(norm(f(x) + J(x)' * y))")
end
println("F  $(norm(c(x)))")
scatter!([x[1]], [x[2]], c=:red, ms=2)
```

```julia
using LinearAlgebra

xg = range(-1.5, 1.5, length=200)
yg = range(-1.5, 1.5, length=200)

# f(x) = x[1]^2 + 4 * x[2]^2
f(x) = x[1]^4 + 4 * x[2]^4
f(x,y) = f([x;y])
# c(x) = [x[1] + x[2] - 1]
c(x) = [x[1]^2 + x[2]^2 - 1]

contour(xg, yg, f, leg=false, levels=50)
g = range(0, 2, length=60)
plot!(cos.(g), sin.(g), c=:black, lw=2)
# plot!(xg, 1 .- xg, c=:black, lw=2)
# g = exp.(range(log(1e-4), log(1e4), length=50))
# x = 4 * g ./ (8 .+ 5g)
# y = g ./ (8 .+ 5g)
# plot!(x, y, c=:red, m=(1))
# scatter!([4/5], [1/5], c=:blue)

 = 1.0
x = [-0.4; 1.0]
y = [0.0]
m = 1

for k = 1:10
    scatter!([x[1]], [x[2]], c=:red, ms=2)
    cx = c(x)
    gx = f(x)
    Ax = J(x)
    Bx = H(x) + sum(Hc(x, i) * (y[i] +  * cx[i]) for i = 1:m)
    x = -(Bx +  * Ax' * Ax) \ (gx + Ax' * (y +  * cx))
    plot!(x[1] .+ [0, x[1]], x[2] .+ [0, x[2]], c=:red, l=:arrow)
    x += x
    y +=  * c(x)
     = max(1.0, 2)
    println("$k  $(norm(c(x)))  $y  $(norm(f(x) + J(x)' * y))")
end
println("F  $(norm(c(x)))")
scatter!([x[1]], [x[2]], c=:red, ms=2)
```
