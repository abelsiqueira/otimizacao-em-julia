```julia
using ForwardDiff
FD = ForwardDiff

∇f(x) = FD.gradient(f, x)
H(x) = FD.hessian(f, x)
J(x) = FD.jacobian(c, x)
Hc(x, i) = FD.hessian(x -> c(x)[i], x)
```

```julia
using LinearAlgebra

xg = range(-1.5, 1.5, length=200)
yg = range(-1.5, 1.5, length=200)

f(x) = x[1]^2 + 4 * x[2]^2
f(x,y) = f([x;y])
# c(x) = [x[1]^2 + x[2]^2 - 1]
c(x) = [x[1] + x[2] - 1]

contour(xg, yg, f, leg=false, levels=50)
θg = range(0, 2π, length=60)
# plot!(cos.(θg), sin.(θg), c=:black, lw=2)
plot!(xg, 1 .- xg, c=:black, lw=2)
# ρg = exp.(range(log(1e-4), log(1e4), length=50))
# xρ = 4 * ρg ./ (8 .+ 5ρg)
# yρ = ρg ./ (8 .+ 5ρg)
# plot!(xρ, yρ, c=:red, m=(1))
# scatter!([4/5], [1/5], c=:blue)

ρ = 1.0
x = [-0.4; 1.0]
m = 1

for k = 1:20
    scatter!([x[1]], [x[2]], c=:red, ms=2)
    cx = c(x)
    gx = ∇f(x)
    Ax = J(x)
    Bx = H(x) + ρ * sum(Hc(x, i) * cx[i] for i = 1:m)
    Δx = -(Bx + ρ * Ax' * Ax) \ (gx + ρ * Ax' * cx)
    plot!(x[1] .+ [0, Δx[1]], x[2] .+ [0, Δx[2]], c=:red, l=:arrow)
    x += Δx
    y = ρ * c(x)
    ρ = max(1.0, 2ρ)
    println("$k  $(norm(cx)) $y  $(norm(∇f(x) + J(x)' * y))")
end
println("F  $(norm(c(x)))")
scatter!([x[1]], [x[2]], c=:red, ms=2)
```

```julia
using LinearAlgebra

xg = range(-1.5, 1.5, length=200)
yg = range(-1.5, 1.5, length=200)

# f(x) = x[1]^2 + 4 * x[2]^2
f(x) = x[1]^4 + 4 * x[2]^4
f(x,y) = f([x;y])
# c(x) = [x[1] + x[2] - 1]
c(x) = [x[1]^2 + x[2]^2 - 1]

contour(xg, yg, f, leg=false, levels=50)
θg = range(0, 2π, length=60)
plot!(cos.(θg), sin.(θg), c=:black, lw=2)
# plot!(xg, 1 .- xg, c=:black, lw=2)
# ρg = exp.(range(log(1e-4), log(1e4), length=50))
# xρ = 4 * ρg ./ (8 .+ 5ρg)
# yρ = ρg ./ (8 .+ 5ρg)
# plot!(xρ, yρ, c=:red, m=(1))
# scatter!([4/5], [1/5], c=:blue)

ρ = 1.0
x = [-0.4; 1.0]
y = [0.0]
m = 1

for k = 1:10
    scatter!([x[1]], [x[2]], c=:red, ms=2)
    cx = c(x)
    gx = ∇f(x)
    Ax = J(x)
    Bx = H(x) + sum(Hc(x, i) * (y[i] + ρ * cx[i]) for i = 1:m)
    Δx = -(Bx + ρ * Ax' * Ax) \ (gx + Ax' * (y + ρ * cx))
    plot!(x[1] .+ [0, Δx[1]], x[2] .+ [0, Δx[2]], c=:red, l=:arrow)
    x += Δx
    y += ρ * c(x)
    ρ = max(1.0, 2ρ)
    println("$k  $(norm(c(x)))  $y  $(norm(∇f(x) + J(x)' * y))")
end
println("F  $(norm(c(x)))")
scatter!([x[1]], [x[2]], c=:red, ms=2)
```

```julia
using Interact

f(x) = (x[1] - 1)^2 + 4 * (x[2] - x[1]^2)^2
# x ≥ 0, x₁ + x₂ ≤ 1
@manipulate for ρ = exp.(range(log(1e-4), log(1e4), length=50))
    ϕ(x) = f(x) + ρ * (
        max(-x[1], 0)^2 + max(-x[2], 0)^2 + max(x[1] + x[2] - 1, 0)^2
    )
    ϕ(x,y) = ϕ([x;y])

    contour(xg, yg, ϕ, leg=false, levels=100)
    # plot!(xg, 1 .- xg, c=:black, lw=2)
end
```
